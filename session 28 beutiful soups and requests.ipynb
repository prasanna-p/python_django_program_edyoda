```python
import requests
from bs4 import BeautifulSoup as bs
import csv

next_link = "/mobiles/pr?sid=tyy,4io&otracker=categorytree"
HOME_URL = "https://www.flipkart.com"
response = requests.get(HOME_URL+next_link)

fp = open("scrap_file2.csv","a+")
writer_obj = csv.DictWriter(fp,lineterminator = "\n",fieldnames = ["name","price","rating","review"])
writer_obj.writerow({"name":"name","price":"price","rating":"rating","review":"review"})

def get_review(card):
    next_link = card.find("a", attrs = {"class":"_31qSD5"})
    response = requests.get(HOME_URL+next_link.get("href"))
    soup = bs(response.content,"html.parser")
    review_content = soup.find_all("p",attrs = {"class":"_2xg6Ul"})
    reviews = [review.text for review in review_content]
    return reviews
    

def get_details(cards):
    for card in cards:
        name = card.find("div",attrs = {"class":"_3wU53n"})
        price = card.find("div",attrs = {"class":"_1vC4OE"})
        rating = card.find("div",attrs = {"class":"hGSR34"})
        review = get_review(card)
        if rating:
            rating = rating.text
        else:
            rating = None
        writer_obj.writerow({"name":name.text,"price":price.text,"rating":rating,"review":review})

count = 50
while count:
    soup = bs(response.content,"html.parser")
    cards = soup.find_all("div", attrs = {"class":"_1UoZlX"})
    get_details(cards)
    count = count-1
    if count:
        print(count)
        next_link = soup.find("a",text = "Next")
        response = requests.get(HOME_URL+next_link.get("href"))
fp.close()
```
